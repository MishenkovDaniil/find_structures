```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.interpolate import make_interp_spline, BSpline

#gaps are caused by other computer processes execution

plt.axis([0,1000000,0,3000])
plt.title('Insert in find structures')
plt.xlabel('size')
plt.ylabel('time, ms')

x = np.array([100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000])
y_insert_avl_rand = np.array([84, 181, 299, 438, 594, 768, 953, 1117, 1302, 1496])
y_insert_avl_sort = np.array([65, 140, 219, 298, 379, 466, 545, 622, 705, 792])
y_insert_treapset_rand = np.array([75, 163, 270, 402, 547, 705, 863, 1030, 1184, 1378])
y_insert_splay_rand = np.array([149, 329, 534, 786, 1067, 1356, 1645, 1925, 2263, 2553])

xnew = np.linspace (x. min (), x. max (), 200 ) 

spl = make_interp_spline (x, y_insert_avl_rand, k = 1)
y_smooth_insert_avl_rand = spl (xnew)
spl = make_interp_spline (x, y_insert_avl_sort, k = 1)
y_smooth_insert_avl_sort = spl (xnew)
spl = make_interp_spline (x, y_insert_treapset_rand, k = 1)
y_smooth_insert_treapset_rand = spl (xnew)
spl = make_interp_spline (x, y_insert_splay_rand, k = 1)
y_smooth_insert_splay_rand = spl (xnew)

plt.plot (xnew, y_smooth_insert_avl_rand, "y",
          xnew, y_smooth_insert_avl_sort, "r",
          xnew, y_smooth_insert_treapset_rand, "g",
          xnew, y_smooth_insert_splay_rand, "m")
plt.legend (["avl on random values", "avl on sort values", "treapset on random values", "splay on random values"])
plt.show()

plt.axis([0,1000000,0,1500])
plt.title('Remove in find structures')
plt.xlabel('size')
plt.ylabel('time, ms')

x = np.array([100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000])
y_remove_avl_rand = np.array([43, 115, 203, 293, 388, 488, 588, 697, 808, 927])
y_remove_avl_sort = np.array([28, 59, 91, 124, 157, 191, 228, 264, 297, 331])
y_remove_treapset_rand = np.array([35, 85, 171, 262, 341, 436, 528, 634, 729, 834])
y_remove_splay_rand = np.array([70, 181, 313, 445, 582, 722, 865, 1012, 1165, 1321])

xnew = np.linspace (x. min (), x. max (), 200 ) 

spl = make_interp_spline (x, y_remove_avl_rand, k = 1)
y_smooth_remove_avl_rand = spl (xnew)
spl = make_interp_spline (x, y_remove_avl_sort, k = 1)
y_smooth_remove_avl_sort = spl (xnew)
spl = make_interp_spline (x, y_remove_treapset_rand, k = 1)
y_smooth_remove_treapset_rand = spl (xnew)
spl = make_interp_spline (x, y_remove_splay_rand, k = 1)
y_smooth_remove_splay_rand = spl (xnew)

plt.plot (xnew, y_smooth_remove_avl_rand, "y",
          xnew, y_smooth_remove_avl_sort, "r",
          xnew, y_smooth_remove_treapset_rand, "g",
          xnew, y_smooth_remove_splay_rand, "m")
plt.legend (["avl on random values", "avl on sort values", "treapset on random values", "splay on random values"])
plt.show()
```
